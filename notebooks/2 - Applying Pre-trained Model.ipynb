```python
from transformers import pipeline
from datasets import load_dataset
```

```python
pipe = pipeline(task='text-classification')
```

```python
ds = load_dataset('imdb')
```

```python
data = ds['test']['text']
```

```python
preds = pipe(data, batch_size=32, truncation=True)
```

```python
y_true = list(ds['test']['label'])
```

```python
label = []
for dic in preds:
    if dic['label'] == 'NEGATIVE':
        label.append(0),
    else:
        label.append(1)
```

```python
probs = [dic['score'] for dic in preds]
```

```python
probs
```

```python
from sklearn.metrics import roc_auc_score
score = roc_auc_score(label, probs)
```

```python
print(score)
```
